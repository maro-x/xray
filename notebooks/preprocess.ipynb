```python
from PIL import Image
from pathlib import Path
import os
import pandas as pd
from glob import glob
```

انا قللت عدد الصور في ال normal بحيث يبقي نفس عدد ال pneumonia وبعدين اشتغلت عادي 

```python

def build_data_map(folder_path):
    img_path = glob(folder_path + r'\*.jpeg')
    clas = Path(folder_path).name
    split = Path(folder_path).parent.name
    return pd.DataFrame({
        'img_path': img_path,
        "clas": clas,
        'split': split
    })
```

```python

training_normal = build_data_map(r"F:\New folder (2)\chest_xray\chest_xray_2\train\NORMAL")
training_sick = build_data_map(r"F:\New folder (2)\chest_xray\chest_xray_2\train\PNEUMONIA")

test_normal = build_data_map(r"F:\New folder (2)\chest_xray\chest_xray_2\test\NORMAL")
test_sick = build_data_map(r"F:\New folder (2)\chest_xray\chest_xray_2\test\PNEUMONIA")

val_normal = build_data_map(r"F:\New folder (2)\chest_xray\chest_xray_2\val\NORMAL")
val_sick = build_data_map(r"F:\New folder (2)\chest_xray\chest_xray_2\val\PNEUMONIA")
```

```python

data = pd.concat([
    training_normal,
    training_sick,
    test_normal,
    test_sick,
    val_normal,
    val_sick,
])
```

```python

width = 128
height = 128
for i, row in data.iterrows():
    img_path = row["img_path"]
    img = Image.open(img_path)
    resized = img.resize((width, height))
    img_path = Path(img_path)
    img_name = img_path.name
    img_clas = img_path.parent.name
    img_split = img_path.parent.parent.name
    final_img_name = img_split + '_' + img_clas + '_' + img_name
    final_img_path = 'C:/Users/DELL/OneDrive/Documents/GitHub/xray/data/resized/' + final_img_name
    resized.save(final_img_path)
```

```python

resized_img_path = []
for i, row in data.iterrows():
    img_path = Path(row["img_path"])
    img_name = img_path.name
    img_clas = img_path.parent.name
    img_split = img_path.parent.parent.name
    final_img_name = img_split + '_' + img_clas + '_' + img_name
    final_img_path = 'C:/Users/DELL/OneDrive/Documents/GitHub/xray/data/resized/' + final_img_name
    resized_img_path.append(final_img_path)
```

```python

data['resized'] = resized_img_path
data.to_csv('C:/Users/DELL/OneDrive/Documents/GitHub/xray/data/data_map.csv', index=False)
```
